```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mouseHandler(event,x,y,flags,param):
    global im_temp, pts_src #image temporaire et les points sélectionnés source.

    if event == cv2.EVENT_LBUTTONDOWN: #si le bouton gauche de la souris est cliqué
        cv2.circle(im_temp,(x,y),3,(0,255,255),5,cv2.LINE_AA) #dessiner un cercle jaune autour du point sélectionné
        #3 : size circle, (0,255,255) : color, 5 : epaisseur de la ligne, cv2.LINE_AA : type de ligne 
        cv2.imshow("Image", im_temp) 
        if len(pts_src) < 4: #si le nombre de points sélectionnés est inférieur à 4, ajouter le point sélectionné à la liste des points source
            pts_src = np.append(pts_src,[(x,y)],axis=0) #ajouter le point sélectionné à la liste des points source



for i in range(0, 4): 
    im_src = cv2.imread("book1.jpg") 
    im_src = cv2.cvtColor(im_src, cv2.COLOR_BGR2RGB)
    # Destination image
    height, width = 400, 600 
    im_dst = np.zeros((height,width,3),dtype=np.uint8) #3: nombre de canaux de couleur (RGB)

    # Create a list of points.
    pts_dst = np.empty((0,2)) #initialiser une liste de points destination vide
    pts_dst = np.append(pts_dst, [(0,0)], axis=0) #ajouter le point (0,0) à la liste des points destination
    pts_dst = np.append(pts_dst, [(width-1,0)], axis=0)
    pts_dst = np.append(pts_dst, [(width-1,height-1)], axis=0)#3: ajouter le point (width-1,height-1) à la liste des points destination
    pts_dst = np.append(pts_dst, [(0,height-1)], axis=0)

    # Create a window
    cv2.namedWindow("Image", 1) # 1 : flag pour la fenêtre, ici on utilise une fenêtre normale

    im_temp = im_src
    pts_src = np.empty((0,2)) #initialiser une liste de points source vide

    cv2.setMouseCallback("Image",mouseHandler) #associer la fonction mouseHandler à la fenêtre "Image" pour gérer les événements de la souris

    cv2.imshow("Image", im_temp)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    tform, status = cv2.findHomography(pts_src, pts_dst)
    im_dst = cv2.warpPerspective(im_src, tform,(width,height)) #appliquer la transformation homographique à l'image source pour obtenir l'image destination

    plt.subplot(1,2,1),plt.imshow(im_src),plt.title('Input')
    plt.subplot(1,2,2),plt.imshow(im_dst),plt.title('Output')
    plt.show()
```
